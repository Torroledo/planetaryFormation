## Data handling 
---
Data are organized in such way that each observation represent a formed planet. We have to collapse data by planetary systems ($ident$). However before do this we should verify that variables don't have variation between systems (no with-in variation). Also, we must create the final variables to predict:

Varibles:

    1. Number of planets.
    2. Planets total mass.
    3. Mass efficiency.

Tasks:
    
    1.Verify no with-in variation.
    2.Create final varaibles.
    3.Collapse data by $ident$. 
    

```python
def collapse_by_system(dataFrame):
    ident_systems = unique(dataFrame['ident'])
    num_systems = len(ident_systems)
    print('There is only '+str(num_systems),'non-empty observations (systems)')
    # Define a new Data Frame to store planetary systems information 
    dataFrame_systems = pd.DataFrame(np.nan, index = range(0,0), columns = dataFrame.columns ) 
    # 1. Getting the # of planets per system
    # 2. Getting the total mass of planets = (emepla/emet)
    # 3. Getting the mass efficiency = total_mass/emed
    num_planets = np.zeros(num_systems)
    solid_mass = np.zeros(num_systems)
    mass_efficiency = np.zeros(num_systems)
    for sys in range(0,num_systems):
        system = ident_systems[sys]
        actual = dataFrame[dataFrame['ident'] == system]
        num_planets[sys] = actual['ident'].count()
        solid_mass[sys] = sum(actual['emepla(i)/emet'])
        mass_efficiency[sys] = solid_mass[sys]/np.min(actual['emed'])
        dataFrame_systems.loc[sys,:] = actual.iloc[0,:]
    # Dropping non-initial conditions
    dataFrame_systems.drop(['emegas(i)','emepla(i)/emet','rplanet(i)/radtie','a(i)'], axis=1, inplace=True)
    dataFrame_systems['num_planets'] = num_planets
    dataFrame_systems['solid_mass'] = solid_mass
    dataFrame_systems['mass_efficiency'] = mass_efficiency
    return dataFrame_systems
```

```python
high_systems = collapse_by_system(high)
```

```python
low_systems = collapse_by_system(low)
```
